```python
num_cols = ['Age','Fare']
cat_cols = ['Sex','SibSp','Pclass','Parch','Embarked']
```

```python
comb = pd.DataFrame(tr)
comb = comb.append(td)
```

```python
comb.isnull().sum()
```

```python
for col in cat_cols:
    comb[col] = comb[col].astype("category")
comb.dtypes
```

```python
comb.drop(['Cabin','Name','Ticket'],axis=1,inplace=True)
comb.columns
```

```python
for col in num_cols:
    comb[[col]] = comb[[col]].fillna(comb.median())
```

```python
comb.isnull().sum()
```

```python
for col in cat_cols:
    comb[[col]] = comb[[col]].fillna(comb.mode().iloc[0])
comb.isnull().sum()
```

```python
import seaborn as sns
grid = sns.FacetGrid(comb,row = 'Embarked' , col = 'Survived' , size = 3.0 , aspect = 1.6)
grid.map(sns.barplot,'Sex','Fare',alpha = .5 , ci = None)
grid.add_legend()
```

```python
labels = ["0","1","2","3"]
comb['AgeBand'] = pd.cut(comb['Age'],4,labels=labels)
comb.head()
```

```python
train = comb.iloc[:891,]
test = comb.iloc[891:,]
print(train.shape,test.shape)
```

```python
test.columns
```

```python
train.drop(['Age'],axis=1,inplace=True)
```

```python
test.drop(['Age','Survived'],axis=1,inplace=True)
```

```python
train.columns
```

```python
test.columns
```

```python
catcols = ['Sex','SibSp','Pclass','Parch','Embarked','AgeBand']
```

```python
train = pd.get_dummies(train,columns = catcols,drop_first=True)
```

```python
test = pd.get_dummies(test,columns = catcols,drop_first = True)
```

```python
train.columns
```

```python
test.columns
```

```python
train['Survived'] = train['Survived'].astype('category').astype('int')
```

```python
X = train.drop(['Survived','PassengerId'],axis=1)
y = train['Survived']
```

```python
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(X,y,test_size =0.3,random_state=42)
print(X_train.shape,X_val.shape,y_train.shape,y_val.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1)
lr.fit(X_train,y_train)
lr_pred = lr.predict(X_val)
```

```python
lr_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score,precision_score,recall_score,f1_score
def print_results(headline,true_value,pred):
    print(headline)
    print('Accuracy : {}'.format(accuracy_score(true_value,pred)))
    print('Precision score: {}'.format(precision_score(true_value,pred)))
    print('Sensitivity: {}'.format(recall_score(true_value,pred)))
    print('F1 score: {}'.format(f1_score(true_value,pred)))
```

```python
print(classification_report(y_val,lr_pred))
```

```python
print_results('Logistic Regression',y_val,lr_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth = 3,min_samples_leaf=30)
dt.fit(X_train,y_train)
dt_pred = dt.predict(X_val)
```

```python
print_results("Decision Tree Classifier",y_val,dt_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,random_state=4)
rf.fit(X_train,y_train)
rf_pred = rf.predict(X_val)
```

```python
print_results("Random Forest Classifier",y_val,rf_pred)
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
param_clf = {'max_depth':range(3,10),
            'criterion':['entropy','gini'],
            'min_samples_leaf':range(20,50,10)
            }
classifier = DecisionTreeClassifier()
dt_grd = GridSearchCV(classifier,param_grid = param_clf,cv = 5)
```

```python
dt_grd.fit(X_train,y_train)
```

```python
print(dt_grd.best_params_)
```

```python
dt_grd_pred = dt_grd.predict(X_val)
```

```python
print_results("Decision Tree GridSearchCV",y_val,dt_grd_pred)
```

```python
classifier = RandomForestClassifier()
param_clf = {'n_estimators': [200,300,400],
            'max_depth' : [200,300,400],
            'min_samples_split':[10,11,12]
            }
rf_rsh = RandomizedSearchCV(classifier,param_clf , cv = 5,random_state = 4)
```

```python
rf_rsh.fit(X_train,y_train)
rf_rsh_pred = rf_rsh.predict(X_val)
```

```python
print_results("Random Forest RandomizedSearch",y_val,rf_rsh_pred)
```

```python
print(rf_rsh.best_params_)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'linear',random_state = 4,probability=True)
svc.fit(X_train,y_train)
svc_pred = svc.predict(X_val)
```

```python
print_results("Support Vector Classifier",y_val,svc_pred)
```

```python
from xgboost import XGBClassifier
xgbc = XGBClassifier(max_depth = 3 ,min_child_weight = 6 ,n_estimators = 250 ,random_state = 42)
xgbc.fit(X_train,y_train)
xgbc_pred = xgbc.predict(X_val)
```

```python
print_results("XGBoost Classifier",y_val,xgbc_pred)
```

```python
print(classification_report(y_val,xgbc_pred))
```
